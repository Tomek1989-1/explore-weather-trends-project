## Explore weather trends

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
warsaw_temp = pd.read_csv('warsaw_results.csv')
warsaw_temp.fillna(warsaw_temp.mean(), inplace=True)
global_temp = pd.read_csv('global_results.csv')
```

```python
warsaw_temp.set_index('year',inplace=True)
global_temp.set_index('year',inplace=True)
```

moving average calculation

```python
warsaw_temp['moving_avg_temp'] = warsaw_temp.avg_temp.rolling(10,min_periods=10).mean()
global_temp['moving_avg_temp'] = global_temp.avg_temp.rolling(10,min_periods=10).mean()
```

data visualization

```python
warsaw_temp.moving_avg_temp.plot(color='red',linewidth=2,figsize=(12,6))
global_temp.moving_avg_temp.plot(color='green',linewidth=2)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.legend(labels =['Warsaw avg temp.', 'Global avg temp.'], fontsize=14)
plt.title('The yearly average air temperature', fontsize=20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Temperature [°C]', fontsize=16);
```

renaming and combining columns into one dataframe to plot a scatterplot

```python
warsaw_temp.rename(columns={'avg_temp':'warsaw_avg_temp','moving_avg_temp':'warsaw_moving_avg_temp'},inplace=True)
global_temp.rename(columns={'avg_temp':'global_avg_temp','moving_avg_temp':'global_moving_avg_temp'},inplace=True)
```

```python
df_combined = warsaw_temp.merge(global_temp, left_on='year', right_on='year', how='inner')
```

```python
df_combined.plot(x='warsaw_moving_avg_temp',y='global_moving_avg_temp',kind='scatter');
```

```python
df_combined.corr()
```
